## 4. Density Weighted Temperature

```python
from yt.units import mp, kb, mh # 從 yt 匯入質子質量和波茲曼常數



def _temperature(field, data):
    mu = 0.6
    # gas_internal_energy_density = data[('boxlib', 'gasInternalEnergy')]
    gas_internal_energy_density = data[('gas', 'internal_energy_density')]
    # gas_density = data[('boxlib', 'gasDensity')]
    gas_density = data[('gas', 'density')]
    temp = 2/3 * gas_internal_energy_density * mu * mh / gas_density / kb
    return temp

ds.add_field(name=('gas', 'temperature'),
             function=_temperature,
             sampling_type="cell",  
             units="K")             # 指定單元為凱氏溫標 (Kelvin)

temperature_densityWeight_prj_plt = yt.ProjectionPlot(ds, 'x', 
                                                       ("gas", "temperature"), 
                                                       weight_field=("gas", "density"))
temperature_densityWeight_prj_plt.show()


internal_energy_slc_plt = yt.SlicePlot(ds, 'x', ('gas', 'internal_energy_density'))
internal_energy_slc_plt.show()  
mass_density_slc_plt = yt.SlicePlot(ds, 'x', ('gas', 'density'))
mass_density_slc_plt.show()
```

# This is same plot but using numpy

```python
import yt
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib.ticker import ScalarFormatter

max_level = ds.max_level
base_dim = ds.domain_dimensions
finest_dims = base_dim * 2**max_level

# --- 1. Get the physical boundaries for the axes ---
# We are slicing along X, so our plot axes will be Y and Z.
y_min, y_max = ds.domain_left_edge.in_units('pc')[1].value, ds.domain_right_edge.in_units('pc')[1].value
z_min, z_max = ds.domain_left_edge.in_units('pc')[2].value, ds.domain_right_edge.in_units('pc')[2].value

# Get the units for labeling the plot
axis_units = 'pc'

grid = ds.covering_grid(level=max_level, 
                        left_edge=ds.domain_left_edge, 
                        dims=finest_dims)

# density
density_yt_array = grid[('gas', 'density')]
print(f'density_yt_array: {density_yt_array.units}')
density_3d_numpy_array = density_yt_array.to_ndarray()

# vz velocity
vz_yt_array = grid[('gas', 'four_velocity_z')]
print(f'vz_yt_array: {vz_yt_array.units}')
vz_3d_numpy_array = vz_yt_array.to_ndarray()
density_weighted_vz_yt_array = density_yt_array * vz_yt_array
print(f'density_weighted_vz_yt_array: {density_weighted_vz_yt_array.units}')
density_weighted_vz_np_array = density_weighted_vz_yt_array.to_ndarray()
density_weighted_vz_np_array_projection = np.sum(density_weighted_vz_np_array, axis=0) / np.sum(density_3d_numpy_array, axis=0)

# temperature
T_yt_array = grid[('gas', 'temperature')]
print(f'temperature_yt_array: {T_yt_array.units}')
temperature_3d_numpy_array = T_yt_array.to_ndarray()  

density_weighted_T = density_yt_array * T_yt_array
density_weighted_T_np_array = density_weighted_T.to_ndarray()
density_weighted_T_np_array_projection = np.sum(density_weighted_T_np_array, axis=0) / np.sum(density_3d_numpy_array, axis=0)


# Define the slice index
x_index = 32
density_2d_slice = density_3d_numpy_array[x_index, :, :]

column_density = np.sum(density_3d_numpy_array, axis=0)

# --- 2. Define the extent based on the transposed array ---
# For density_2d_slice.T:
# Horizontal axis is Y, Vertical axis is Z
# extent = [horizontal_min, horizontal_max, vertical_min, vertical_max]
plot_extent = [y_min, y_max, z_min, z_max]

#################################################################################
# --- 3. Create the plot with the correct extent and labels ---
fig, ax = plt.subplots(figsize=(6, 10))

# Use the transposed slice to make Z the vertical axis
im = ax.imshow(density_2d_slice.T, 
               origin='lower',
               extent=plot_extent,  # <-- Use the physical extent here
               aspect='auto',
               norm=LogNorm(), 
               cmap='viridis')

# Add units to the labels
fig.colorbar(im, ax=ax, label=f'Density ({density_yt_array.units})')
ax.set_title(f'Density Slice at x-index {x_index}')
ax.set_xlabel(f'Y ({axis_units})')
ax.set_ylabel(f'Z ({axis_units})')

plt.savefig('np_density_slice_corrected.png', dpi=300)
plt.show()
##########################################################################################

##########################################################################################
# --- 3. Create the plot with the correct extent and labels ---
fig, ax = plt.subplots(figsize=(6, 10))

# Use the transposed slice to make Z the vertical axis
im = ax.imshow(column_density.T, 
               origin='lower',
               extent=plot_extent,  # <-- Use the physical extent here
               aspect='auto',
               norm=LogNorm(), 
               cmap='viridis')

# Add units to the labels
fig.colorbar(im, ax=ax, label=f'Density ({density_yt_array.units})')
ax.set_title(f'Density Projection')
ax.set_xlabel(f'Y ({axis_units})')
ax.set_ylabel(f'Z ({axis_units})')

plt.savefig('np_density_projection_corrected.png', dpi=300)
plt.show()
######################################################


##########################################################################################
# --- 3. Create the plot with the correct extent and labels ---
fig, ax = plt.subplots(figsize=(6, 10))

# Use the transposed slice to make Z the vertical axis
im = ax.imshow(density_weighted_vz_np_array_projection.T, 
               origin='lower',
               extent=plot_extent,  # <-- Use the physical extent here
               aspect='auto',
               cmap='viridis')

formatter = ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((0, 0)) # 這會強制對所有數量級的數字都使用科學記數法
# Add units to the labels
cbar = fig.colorbar(im, ax=ax, label=f'Density Weighted Velocity-z ({density_yt_array.units})', format=formatter)
ax.set_title(f'Density Weighted vz Projection')
ax.set_xlabel(f'Y ({axis_units})')
ax.set_ylabel(f'Z ({axis_units})')

plt.savefig('np_density_weighted_vz_projection_corrected.png', dpi=300)
plt.show()
######################################################



##########################################################################################
# --- 3. Create the plot with the correct extent and labels ---
fig, ax = plt.subplots(figsize=(6, 10))

# Use the transposed slice to make Z the vertical axis
im = ax.imshow(density_weighted_T_np_array_projection.T, 
               origin='lower',
               norm=LogNorm(),
               extent=plot_extent,  # <-- Use the physical extent here
               aspect='auto',
               cmap='viridis')

formatter = ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((0, 0)) # 這會強制對所有數量級的數字都使用科學記數法
# Add units to the labels
cbar = fig.colorbar(im, ax=ax, label=f'Density Weighted T ({T_yt_array.units})', format=formatter)
ax.set_title(f'Density Weighted T Projection')
ax.set_xlabel(f'Y ({axis_units})')
ax.set_ylabel(f'Z ({axis_units})')

plt.savefig('np_density_weighted_T_projection_corrected.png', dpi=300)
plt.show()
######################################################

```
