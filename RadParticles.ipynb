```python
import yt

# Load your dataset
ds = yt.load("sample/RadiatingParticles/plt026")

# 1. Create a base plot. A SlicePlot is a good choice.
# We'll plot along the 'z' axis and show the radiation energy density.
plot = yt.SlicePlot(ds, "z", ("rad", "energy_density_0"))

# 2. Use annotate_quiver() to overlay the vectors.
# For a slice along the 'z' axis, you need to provide the x and y flux components.
plot.annotate_quiver(
    ("rad", "flux_density_x_0"),
    ("rad", "flux_density_y_0"),
    factor=15,  # You can adjust this number to change the length of the arrows
    normalize=False  # Normalize the vectors for better visualization
)

# 3. Customize and save the plot

plot.set_log(("rad", "energy_density_0"), True) # Use a log scale for better visualization
plot.save("radiation_flux_quiver.png")
plot.show()
# To display in a notebook, you can just call the plot object
# plot
```

```python
import yt

ds = yt.load("sample/RadiatingParticles/plt026")

# 1. 繪製 X 方向的通量
plot_x = yt.SlicePlot(ds, "z", ("rad", "flux_density_x_0"))
plot_x.set_cmap(("rad", "flux_density_x_0"), "RdBu_r") # 使用一個能顯示正負值的 colormap

plot_x.save("flux_component_x.png")

# 2. 繪製 Y 方向的通量
plot_y = yt.SlicePlot(ds, "z", ("rad", "flux_density_y_0"))
plot_y.set_cmap(("rad", "flux_density_y_0"), "RdBu_r")

plot_y.save("flux_component_y.png")

# 在 Notebook 中顯示
plot_x.show()
plot_y.show()
```
